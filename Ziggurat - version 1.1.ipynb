```python
#############################################################################################

#inputSet = measure settings list
#measures = measureOlist obtained in the read plant file function previously
#tagSet = selected tags, this is a required input (with selected measures s)

def configs(inputSet, measures, tagSet): 
    
    workshopTags = []
    selectedIndex = []

    print("\n\t\t\t\tFILES:\n")
    
    #open tags file
    tagsFile = open(tagsPath, 'r')
    tagsRows = tagsFile.readlines()
    
    maxSet = len(inputSet)
    maxTag = len(tagSet)

    for i in range(0,maxSet):
        index = inputSet[i]
        selectedIndex.append(measures[index].strip()) #measures list with set points
        print(measures[index].strip())

    print("\n\t\t\t\tTAGS")

    for i in range (0,maxTag):
        workshopTags.append(tagsRows[tagSet[i]].strip()) #tag list from setings list
        print(tagsRows[tagSet[i]].strip())
        
    return workshopTags, selectedIndex

############################################################################################

    
workshop1, index1 = configs(input1, plant1Measures, tag1)
workshop2, index2 = configs(input2, plant2Measures, tag2)

```

```python
#Function to count rows of a file
def count(fname):
    return sum(1 for line in open(fname))

import xlwt

#INPUTS#
#inputSet = list of selected measures
#tagSet = list of selected tags
#measures = list measureOlist, which contains only the selected measure file names
#rawFolder = raw data path folder (raw 1 or raw2)
#selected = selected index (result of configs function)
#tags = workshop tags (result of configs function)

#############################################################################################
def batchGenerator(inputSet, rawFolder, selected, tags, results):
    
    maxSet = len(inputSet)
    
    #Lists initialization
    measureDay = [] #consumption of a day (by hours) records are saved in this list
    measureFP = 1 #column increment
    soma = 0 #the sum of consumption in a day is recorded in this variable
    sumDay = [] #final product: [day1, sum of the day1, day2, sum of the day2 ...]
    numLin = 0
    plant = []
    
    rowsCount = 0

    wb = xlwt.Workbook()
    ws = wb.add_sheet('Batch Data')


    for i in range (0, maxSet):
    
        path = rawFolder.strip() + "/" + selected[i]
        file = open(path, 'r') #open file

        print("file name:" , selected[i])
        
        plant.append(tags[i])
        
        numLin = count(path)
        fileList = file.readlines() #list of that file opened
        
        dates = fileList[17] #days registered in that file opened
        datesList = dates.split() #split the recorded row in strings (each string is a date)
        
        numDays = len(datesList) #number of strings splited = number of days in the opened file
        print("days in this file: " , numDays)

        for j in range(0,numDays):

            measureDay.append(datesList[j]) #add date in the first element of the list
            sumDay.append(datesList[j]) #add date in the first element of the list with consumption sum

            for k in range(19, numLin): #row 19 is the first line with measures
                infos = fileList[k] #put the currently row in a list
                infoList = infos.split("\t") #split in \t

                transition = float(infoList[measureFP]) #record the measure in the referred colum (every loop needs an increment of 2 in the column to synchronize with the currently day (j)) 
                measureDay.append(transition) #record in the day measure list the currently measure
                soma = soma + transition

            measureFP = measureFP + 2
            sumDay.append("%.2f" % soma) #the sum result is recorded in the second element of the list sumDay


            ws.write(rowsCount, 2, tags[i]) #column 2
            ws.write(rowsCount, 3, soma) #column 4
            ws.write(rowsCount, 0, datesList[j]) #column 1

            soma = 0

            rowsCount = rowsCount + 1


        wb.save(results.strip() + '/batch.xls')

        plant.append(sumDay)
        
        #reset lists and variables for a new day loop
        measureDay = [] 
        infos = []
        infoList = []
        transition = ""
        sumDay = []
        measureFP = 1
        
    return plant
###########################################################################################       

#Batch for plant 1 and plant 2 files: [fileName, DAY LIST]
plant1 = batchGenerator(input1, raw1, index1, workshop1, result1)
plant2 = batchGenerator(input2, raw2, index2, workshop2, result2)
```

```python

def toFloat(listStr):
    n = len(listStr)
    for i in range(1,n,2):
        listStr[i] = round(float(listStr[i]), 2)

    return listStr

def addAreaDay(areaConsumption, result):
    for i in range(1,len(areaConsumption), 2):
        result[i] = round(result[i],2) + round(areaConsumption[i],2)
    return result

num = len(plant1)
num2 = len(plant2)

#for i in range (0, num):
    #print(plant1[i])

max1 = len(plant1)
max2 = len(plant2)

#tag Names positions

print("\nPlant 1 - Positions\n")

for i in range(0,max1,2):

  print(i,":" , plant1[i])

print("\nPlant 2 - Positions\n")

for i in range(0,max2,2):
  print(i, ":", plant2[i])

#sum without foundry - Plant1
plant1sum = []

withoutFoundry = []

batch = plant1[1] #compressor
plant1sum = toFloat(batch)
batch = toFloat(plant1[3]) #labs
plant1sum = addAreaDay(batch, plant1sum) #labs + compressor
batch = toFloat(plant1[11]) #mbu2
plant1sum = addAreaDay(batch, plant1sum) #labs + compressor + mbu2
batch = toFloat(plant1[17]) #maintenance/tooling
plant1sum = addAreaDay(batch, plant1sum) #labs + compressor + mbu2 + maintenance/tooling

for i in range(0,len(plant1sum)):
    withoutFoundry.append(plant1sum[i])

batch = toFloat(plant1[9]) #mbu1
plant1sum = addAreaDay(batch, plant1sum) #total with foundry

print("Plant 1 w/ Foundry:", withoutFoundry)
print("Plant 1 Total: " , plant1sum)

plant2sum = []

batch = plant2[3] #mbu3
plant2sum = toFloat(batch)
batch = toFloat(plant2[5]) #mbu4
plant2sum = addAreaDay(batch, plant2sum)
batch = toFloat(plant2[7]) #mbu5
plant2sum = addAreaDay(batch, plant2sum)
batch = toFloat(plant2[9]) #compressor
plant2sum = addAreaDay(batch, plant2sum) 
batch = toFloat(plant2[21]) #ADM
plant2sum = addAreaDay(batch, plant2sum)
batch = toFloat(plant2[23]) #RH
plant2sum = addAreaDay(batch, plant2sum)

print("Plant 2:" , plant2sum)

listLenght = len(plant1sum)

tecumsehWithoutFoundry = []
tecumseh = []

for i in range(0, listLenght):
    tecumsehWithoutFoundry.append(withoutFoundry[i]) #without foundry
    tecumseh.append(plant1sum[i]) #with foundry

#tecumseh total

for i in range(1, listLenght,2):
    tecumsehWithoutFoundry[i] = tecumsehWithoutFoundry[i] + plant2sum[i]
    tecumseh[i] = tecumseh[i] + plant2sum[i]
    
    
print("total Tecumseh:", tecumseh)
print("without Foundry:" , tecumsehWithoutFoundry)


```
