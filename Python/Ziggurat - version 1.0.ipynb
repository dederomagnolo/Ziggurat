```python
#ONE FILE SCRIPT - RETURNS NUMBER OF DAYS AND SUM

#Function: Count lines of an archive

#row 17 = dates line: count all dates in this line to know how many measures will be saved
def count(fname):
    return sum(1 for line in open(fname))


#To do: rotina para ler todos os arquivos de uma pasta


#Open archives
fileName = "C:/Users/debora.romagnolo/Desktop/Python/Medições Virtuais_[6] MBU2.txt"
#archiveName = "SMART BASE 1_[8] FORÇA ELETRO MOTRIZ FUND. 1.txt"
file = open(fileName, 'r')
archiveList = file.readlines() #transform the readed txt file in a list
file.close()

# Info from the opened archive

linFP = archiveList[19] #first record "fora de ponta"
linP = archiveList[38] #rows 38,39,40 are "ponta" records

listFP = linFP.split() #split row in strings
listP = linP.split()

lenFP = len(listFP)
lenP = len(listP)

print("================================================= INFORMAÇÕES ==================================================")

print("\nArquivo:" , fileName) 
print("\nNúmero de linhas no arquivo:" , count(fileName))
print("Número de elementos quando fora de ponta:" , lenFP)
print("Número de elementos quando na ponta:" , lenP)

dates = archiveList[17] #record the dates row

datesList = dates.split() #split the recorded row in strings (each string is a date)
numDays = len(datesList) #number of strings splited = number of days in the opened file

#row 18 = measure titles
#row 19 = first line of records 

iterations = count(fileName) - 18 #iterations number required
numLin = count(fileName) #max number of rows in the opened file

print("Iterações do script:" , iterations , "\n") #print number of iterations of the script for the opened file

print("=============================================== DATAS ENCONTRADAS =============================================== ")
print("\n\t\t\t\tNo relatório lido foram encontradas" , numDays , "datas!\n") #indica o número de datas no arquivo lido
print(dates)
print("================================================================================================================= ")

#Collect data from the opened file
#The product of this routine is a list with measures of a day (by hours) and the consumption sum of each day.

#Lists initialization
measureDay = [] #consumption of a day (by hours) records are saved in this list
measureFP = 1 #column increment
soma = 0 #the sum of consumption in a day is recorded in this variable
sumDay = [] #final product: [day1, sum of the day1, day2, sum of the day2 ...]

#the final product would have 24 measures + the currently day = 25 rows

for j in range(0,numDays):
    
    measureDay.append(datesList[j]) #add date in the first element of the list
    sumDay.append(datesList[j])
    
    for i in range(19, numLin): #row 19 is the first line with measures
        infos = archiveList[i] #put the currently row in a list
        infoList = infos.split("\t") #split in \t

        transition = float(infoList[measureFP]) #record the measure in the referred colum (every loop needs an increment of 2 in the column to synchronize with the currently day (j)) 
        measureDay.append(transition) #record in the day measure list the currently measure
        soma = soma + transition
    
    measureFP = measureFP + 2
    sumDay.append("%.2f" % soma) #the sum result is recorded in the second element of the list sumDay
    
    print(sumDay) #show results
    
    #print("%.2f" % sumDay) #apresentar soma arredondando duas casas
    #print(measureDay)
    
    #reset lists and variables for a new day loop
    measureDay = [] 
    infos = []
    infoList = []
    transition = ""

```
